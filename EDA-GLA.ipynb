```python
def createTarget(dataset):
    target = np.arange(1, dataset.shape[0] + 1)
    target = pd.Series(target)
    return target
    
def addTargetCol(target, dataset):
    full_dataset = dataset
    full_dataset.insert(dataset.shape[1], "popularity", target)
    return full_dataset


target_dataset_2017 = createTarget(dataset_2017)
target_dataset_2018 = createTarget(dataset_2018)

revised_dataset_2017 = addTargetCol(target_dataset_2017, dataset_2017)
revised_dataset_2018 = addTargetCol(target_dataset_2017, dataset_2018)
```

```python
def getCorrelation(dataset):
    import pandas as pd
    corr = dataset.corr('pearson')['popularity']
    corr_sq = corr.pow(2)
    corr = pd.DataFrame({"variables": list(dataset.columns),
                         "corr_score": corr.to_numpy(),
                         "corr_sq": corr_sq.to_numpy()})
    corr = corr.sort_values(by=['corr_sq'], ascending=False)

    return pd.DataFrame({"variables": corr['variables'].to_numpy(),
                         "corr_score": corr['corr_score'].to_numpy()})
dataset_reduced = dataset_big.drop(['index', 'name', 'artists'], axis=1)
print(getCorrelation(dataset_reduced))
dataset_reduced.insert(0, 'bias', 1)
print(dataset_reduced)
```

```python
from sklearn.model_selection import train_test_split
X = dataset_reduced.drop(['popularity', 'spch', 'acous', 'dur', 'nrgy', 'live'], axis=1)
y = dataset_reduced.popularity
print(X)
print(y)
low = 1000
lowindex = 0
diff = []
m_res = []
m_pred = []

for i in range(10000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=i)
    
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    scaler.fit(X_train)
    scaler.transform(X_train)
    scaler.transform(X_test)
    
    from sklearn import linear_model
    reg = linear_model.LinearRegression()
    reg.fit(X_train, y_train)
    #print("Weights: ", reg.coef_)
    y_pred = reg.predict(X_test)
    #print("y_pred: ", y_pred)
    pred_error = np.abs(y_test.to_numpy() - y_pred)
    
    from sklearn.metrics import mean_squared_error
    rmse = mean_squared_error(y_test, y_pred)
    #print("RMSE: %d" % i, rmse)
    if rmse < low:
        low = rmse
        lowindex = i
        diff = pred_error

# Append best results from linear model
m_pred.append((np.mean(diff), np.min(diff), np.max(diff), np.std(diff)))
        
print("Low Index: %d, Low RMSE: %d" % (lowindex, low))

def runningAverage(values):
    ra = []
    for i in range(len(values)):
        ra.append(values[:i].mean())
    return ra

runavg = runningAverage(diff)
fig3 = plt.figure(figsize=(15, 15))
ax_1 = fig3.add_subplot(336)
ax_1.set_xlabel("Entry")
ax_1.set_ylabel("Difference")
ax_1.set_title("Entry vs Difference %d" % lowindex)
ax_1.plot(diff)
ax_2 = fig3.add_subplot(336)
ax_2.plot(runavg)
#ax.set_ylim(ax.get_ylim()[::-1])
```

```python
# Artificial Neural Network Regression
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=i)
scaler = StandardScaler()
scaler.fit(X_train)
scaler.transform(X_train)
scaler.transform(X_test)
```

```python
# basic model 1
def basic_model():
    model = Sequential()
    model.add(Dense(7, input_dim=7, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

est = KerasRegressor(build_fn=basic_model, epochs=100, batch_size=8, verbose=1)
kfold = KFold(n_splits=5)
results = cross_val_score(est, X_train, y_train, cv=kfold)
m_res.append((int(abs(results.mean())), int(results.std())))
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))

est.fit(X_train, y_train)
y_predict = est.predict(X_test)
pred_error = np.abs(y_test - y_predict)
m_pred.append((np.mean(pred_error), np.min(pred_error), np.max(pred_error), np.std(pred_error)))
```

```python
# deeper model 2
def deeper_model():
    model = Sequential()
    model.add(Dense(7, input_dim=7, kernel_initializer='normal', activation='relu'))
    model.add(Dense(3, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

est = KerasRegressor(build_fn=deeper_model, epochs=100, batch_size=8, verbose=1)
kfold = KFold(n_splits=5)
results = cross_val_score(est, X_train, y_train, cv=kfold)
m_res.append((int(abs(results.mean())), int(results.std())))
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))

est.fit(X_train, y_train)
y_predict = est.predict(X_test)
pred_error = np.abs(y_test - y_predict)
m_pred.append((np.mean(pred_error), np.min(pred_error), np.max(pred_error), np.std(pred_error)))
```

```python
# wider model 3
def wider_model():
    model = Sequential()
    model.add(Dense(16, input_dim=7, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

est = KerasRegressor(build_fn=wider_model, epochs=100, batch_size=8, verbose=1)
kfold = KFold(n_splits=5)
results = cross_val_score(est, X_train, y_train, cv=kfold)
m_res.append((int(abs(results.mean())), int(results.std())))
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))

est.fit(X_train, y_train)
y_predict = est.predict(X_test)
pred_error = np.abs(y_test - y_predict)
m_pred.append((np.mean(pred_error), np.min(pred_error), np.max(pred_error), np.std(pred_error)))
```

```python
# wide and deep model 4
def wd_model():
    model = Sequential()
    model.add(Dense(16, input_dim=7, kernel_initializer='normal', activation='relu'))
    model.add(Dense(7, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

est = KerasRegressor(build_fn=wd_model, epochs=100, batch_size=8, verbose=1)
kfold = KFold(n_splits=5)
results = cross_val_score(est, X_train, y_train, cv=kfold)
m_res.append((int(abs(results.mean())), int(results.std())))
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))

est.fit(X_train, y_train)
y_predict = est.predict(X_test)
pred_error = np.abs(y_test - y_predict)
m_pred.append((np.mean(pred_error), np.min(pred_error), np.max(pred_error), np.std(pred_error)))
```

```python
# insane model 5
def insane_model():
    model = Sequential()
    model.add(Dense(32, input_dim=7, kernel_initializer='normal', activation='relu'))
    model.add(Dense(16, kernel_initializer='normal', activation='relu'))
    model.add(Dense(8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

est = KerasRegressor(build_fn=insane_model, epochs=1000, batch_size=16, verbose=1)
kfold = KFold(n_splits=5)
results = cross_val_score(est, X_train, y_train, cv=kfold)
m_res.append((int(abs(results.mean())), int(results.std())))
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))

est.fit(X_train, y_train)
y_predict = est.predict(X_test)
pred_error = np.abs(y_test - y_predict)
m_pred.append((np.mean(pred_error), np.min(pred_error), np.max(pred_error), np.std(pred_error)))
```
